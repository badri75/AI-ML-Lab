```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Linear Regression\DS1.xls",engine="xlrd")
df
```

```python
x_mean = df["X"].mean()
y_mean = df["Y"].mean()
print(x_mean,y_mean)
```

```python
df["XY"] = df["X"] * df["Y"]
xy_mean = df["XY"].mean()
print(xy_mean)
```

```python
df["X^2"] = df["X"] * df["X"]
x2_mean = df["X^2"].mean()
x2_mean
```

```python
a1 = (xy_mean - (x_mean * y_mean)) / (x2_mean -(x_mean * x_mean))
a0 = y_mean - (a1 * x_mean)
a0, a1
```

```python
y_pred = a0 + a1 *df["X"]
df["Y_Pred"] = y_pred
df
```

```python
#Finding Regression (R^2)
a = df["Y_Pred"] - y_mean
b = df["Y"] - y_mean
a2 = a*a
b2 = b*b
a2_sum = a2.sum()
b2_sum = b2.sum()
r2 = a2_sum / b2_sum
r2
```

```python
import math
x = df["Y"] - df["Y_Pred"]
x2 = x*x
y = x2.mean()
rmse = math.sqrt(y)
rmse
```
